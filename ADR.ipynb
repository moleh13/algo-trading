```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
start = "2024-01-01"
end = "2024-03-19"
ticker = "ARB11841-USD"
interval = "1d"
pd.options.display.float_format = '{:.4f}'.format
```

```python
df = yf.download(ticker, start=start, end=end, interval=interval)
df
```

```python
df['Daily Range'] = df['High'] - df['Low']
df
```

```python
average_daily_range = df['Daily Range'].mean()

print(f"The average daily range for {ticker} is {average_daily_range:.2f}")
```

```python
# Create figure and subplots
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Daily Range on the first subplot
ax1.plot(df.index, df['Daily Range'], color='blue', linewidth=1, label='Daily Range')
ax1.set_xlabel('Date')
ax1.set_ylabel('Daily Range', color='blue')
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper left')

# Create a second y-axis for the Close prices
ax2 = ax1.twinx()
ax2.plot(df.index, df['Close'], color='red', linewidth=1, label='Close Price')
ax2.set_ylabel('Price', color='red')
ax2.tick_params('y', colors='red')
ax2.legend(loc='upper right')

# Title
plt.title(f"{ticker} Daily Range and Price")

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```
