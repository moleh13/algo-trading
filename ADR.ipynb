```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
start = "2024-01-01"
end = "2024-03-19"
ticker = "CRV-USD"
interval = "1d"
pd.options.display.float_format = '{:.4f}'.format
```

```python
df = yf.download(ticker, start=start, end=end, interval=interval)
df
```

```python
df['Daily Range'] = df['High'] - df['Low']
df
```

```python
average_daily_range = df['Daily Range'].mean()

print(f"The average daily range for {ticker} is {average_daily_range:.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Daily Range'], color= 'blue', linewidth=1, label='Daily Range')
plt.title(f"{ticker} Daily Range")
plt.xlabel("Date")
plt.ylabel("Daily Range")
plt.legend()
plt.grid(True)
plt.show()
```
