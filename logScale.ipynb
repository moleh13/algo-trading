```python
# Import yfinance, matplotlib, and numpy
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Download the data for the BTC-USD pair
btc = yf.download('BTC-USD', start='2015-01-01', end='2024-03-19')

#Calculate the 20 day moving average
btc['20d'] = btc['Close'].rolling(window=20).mean()

#Calculate the 50 day moving average
btc['50d'] = btc['Close'].rolling(window=50).mean()

#Create plot of log scale of the closing price
plt.figure(figsize=(10, 7))
plt.plot(btc['Close'], label='BTC-USD')
plt.yscale('log')
plt.title('BTC-USD')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
#Fit a trend tunnel for this graph
plt.figure(figsize=(10, 7))
plt.plot(btc['Close'], label='BTC-USD')
plt.plot(btc['20d'], label='20d')
plt.plot(btc['50d'], label='50d')
plt.yscale('log')
plt.title('BTC-USD')
plt.xlabel('Date')
```
